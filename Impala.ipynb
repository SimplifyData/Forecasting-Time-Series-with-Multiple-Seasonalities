```python
import pandas as pd
import numpy as np
```

### Fixing Columns
###### The column headings are incorrect

```python
columns = prod_df.iloc[0].str.strip().str.lower().str.replace("\n","_").str.replace(" ","_").tolist()
```

##### Features names created

```python
columns
```

### Replacing Feature names

```python
df_col = {}

for i,col in enumerate(prod_df.columns):
    df_col[col] = columns[i]

prod_df = prod_df.rename(df_col, axis = "columns")

```

```python
df_col
```

### Dropping columns names which was a row before

```python
prod_df.drop(prod_df.index[0], inplace= True)
```

```python
prod_df.head()
```

### Checking and cleaning Status -  
#### Dropping abandoned wells

```python
prod_df.status.unique()
```

```python
prod_df.status = prod_df.status.str.lower().str.replace(" ","_")
```

##### Names of the well status

```python
prod_df.status.unique()
```

```python
prod_df.status = prod_df.status.replace("abandoned,_re-entered_gas", "abandoned_re-entered_gas")
```

##### Cleaned Well Status names

```python
prod_df.status.unique()
```

```python
prod_df.head()
```

```python
abd_wells = prod_df.status[prod_df.status.str.contains("abandoned", regex = False)].unique().tolist()
```

#### All The abandoned well statuses

```python
abd_wells
```

```python
prod_df[prod_df.status.isin(abd_wells)].status.unique()
```

##### Removing abandoned wells

```python
abd_well_index = prod_df[~prod_df.status.isin(abd_wells)].index
```

```python
prod_df = prod_df[~prod_df.status.isin(abd_wells)].reset_index()
```

```python
prod_df.head()
```

```python
prod_df.status.unique()
```
